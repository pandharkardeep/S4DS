# LazyPredict

```python
df.head()

```

```python
from sklearn.preprocessing import LabelEncoder
df['Date_of_Surgery']=pd.to_datetime(df['Date_of_Surgery'])
df['Date_of_Last_Visit'] = pd.to_datetime(df['Date_of_Last_Visit'])
m = {'Negative':0,'Positive':1}
df['ER status'] = df['ER status'].map(m)
df['PR status'] = df['PR status'].map(m)
df['HER2 status'] = df['HER2 status'].map(m)
m = {'FEMALE':1,'MALE':0}
df['Gender'] = df['Gender'].map(m)
le1 = LabelEncoder()
df['Histology'] = le1.fit_transform(df['Histology'])
le2 = LabelEncoder()
df['Surgery_type'] = le2.fit_transform(df['Surgery_type'])
df.head()
le3 = LabelEncoder()
df['Tumour_Stage'] = le3.fit_transform(df['Tumour_Stage'])
df.head()
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.3,random_state =123)

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)

models,predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```

Observations:
1. We can see that Decision Tree Classifier is best model, as it has highest accuracy as well as time taken is least.
2. Quadratic Discriminant Analysis is worst Model, as it has least accuracy of all.

```python
from sklearn.tree import DecisionTreeClassifier
classifier2=DecisionTreeClassifier(max_depth=40)
classifier2.fit(X_train.drop(['Date_of_Surgery','Date_of_Last_Visit'],axis = 1),y_train)

y_pred = classifier2.predict(X_test.drop(['Date_of_Surgery','Date_of_Last_Visit'],axis = 1))
```

```python
pd.Series(y_pred).value_counts()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

We can once again conclude that Decision Tree is the best Classification for our particular Dataset
